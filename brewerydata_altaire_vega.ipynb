## 2 — Altair (2 charts)

```python

import altair as alt

#enable altair
alt.renderers.enable('default')
```

```python

# Chart 1: # of breweries by state & brewery type

chart_state_type = (
    alt.Chart(breweries)
    .mark_bar()
    .encode(
        x=alt.X('state:N', sort='-y', title='State'),
        y=alt.Y('count():Q', title='Number of Breweries'),
        color=alt.Color('brewery_type:N', title='Brewery Type'),
        tooltip=[
            alt.Tooltip('state:N', title='State'),
            alt.Tooltip('brewery_type:N', title='Type'),
            alt.Tooltip('count():Q', title='Brewery Count')
        ]
    )
    .properties(
        width=500,
        height=300,
        title='Number of Breweries by State and Type'
    )
)

chart_state_type
```

```python

# Chart 2: Brewery locations by latitude/longitude

chart_locations = (
    alt.Chart(breweries)
    .mark_circle(opacity=0.7)
    .encode(
        x=alt.X('longitude:Q', title='Longitude'),
        y=alt.Y('latitude:Q', title='Latitude'),
        color=alt.Color('brewery_type:N', title='Brewery Type'),
        tooltip=[
            alt.Tooltip('name:N', title='Brewery'),
            alt.Tooltip('brewery_type:N', title='Type'),
            alt.Tooltip('city:N', title='City'),
            alt.Tooltip('state:N', title='State'),
            alt.Tooltip('website_url:N', title='Website')
        ]
    )
    .properties(
        width=500,
        height=300,
        title='Brewery Locations by Type'
    )
)

chart_locations
```


## 3 - Interaction (tooltips + cross-filter)



```python

# Cross-filtering with Altair: select state(s) to filter brewery types

# Multi-select on 'state'
state_select = alt.selection_multi(fields=['state'], name='StateSelect')

# Interactive chart 1: States with selection
interactive_states = (
    alt.Chart(breweries)
    .mark_bar()
    .encode(
        x=alt.X('state:N', sort='-y', title='State'),
        y=alt.Y('count():Q', title='Number of Breweries'),
        color=alt.condition(
            state_select,
            alt.Color('state_size_bin:N', title='State Size Bin'),
            alt.value('lightgray')
        ),
        tooltip=[
            alt.Tooltip('state:N', title='State'),
            alt.Tooltip('count():Q', title='Brewery Count'),
            alt.Tooltip('state_size_bin:N', title='State Size Category')
        ]
    )
    .add_params(state_select)
    .properties(
        width=500,
        height=300,
        title='Select States to Filter Brewery Types'
    )
)

# Interactive chart 2: Brewery types filtered by selected states
interactive_types = (
    alt.Chart(breweries)
    .mark_bar()
    .encode(
        x=alt.X('brewery_type:N', title='Brewery Type'),
        y=alt.Y('count():Q', title='Number of Breweries'),
        color=alt.Color('brewery_type:N', legend=None, title='Brewery Type'),
        tooltip=[
            alt.Tooltip('brewery_type:N', title='Brewery Type'),
            alt.Tooltip('count():Q', title='Brewery Count')
        ]
    )
    .transform_filter(state_select)
    .properties(
        width=500,
        height=300,
        title='Brewery Types in Selected State(s)'
    )
)

interactive_states | interactive_types
```
